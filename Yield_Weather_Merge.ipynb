Concantenate all province weather data into one dataframe

```python
import glob
```

```python
import pandas as pd
```

### Read all csv files in a folder

```python
path = 'D:\Thesis\Datasets\Annual CSV Files'
csv_files = glob.glob(path+ "\*.csv" )
```

```python
df_list = (pd.read_csv(file) for file in csv_files)
```

```python
weather_data = pd.concat(df_list,ignore_index=True)
```

```python
weather_data
```

### Remove 2022 and 2023 rows

```python
weather_data['year'].unique()
```

```python
years = [2022,2023]
weather_data = weather_data[weather_data.year.isin(years)==False]
```

```python
weather_data=weather_data.drop(['Unnamed: 0'],axis =1)
```

```python
weather_data.rename(columns = {'year':'Year'},inplace = True)
```

```python
weather_data = weather_data.reset_index(drop=True)
```

```python
weather_data
```

### Merge Yield and Weather 

```python
yield_data = pd.read_csv(r"D:\Thesis\Mindanao Dataset.csv") 
```

```python
yield_data
```

```python
yield_data=yield_data.drop(['Unnamed: 0'],axis =1)
```

```python
yield_data
```

## Merge weather and yield dataset on common year and location

Uncommon year and location will not be included in the final dataset. This implies that locations who have weather data for certain years but do not have yield data for those years (have not cultivated robusta coffee yet) are removed. Example: Zamboanga Sibugay has weather data from 1990 - 2021 but only cultivated coffee starting from year 2000 so only the records for year 2000 and above are included in the final dataset

```python
yield_weather = pd.merge(weather_data,yield_data, how = 'inner', on =['Year','Location'])
```

```python
yield_weather
```

### Sample result for Zamboanga Sibugay

```python
yield_weather[yield_weather['Location'] == 'ZAMBOANGA SIBUGAY']
```

### Categorical Encoding of Wind Direction

```python
from sklearn import preprocessing 
```

```python
label_encoder = preprocessing.LabelEncoder() 
```

```python
yield_weather['Wind_Direction']= label_encoder.fit_transform(yield_weather['Wind_Direction']) 
```
