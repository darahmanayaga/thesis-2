# Get the list of official province names from https://psgc.gitlab.io/api/ 

Philippine Standard Geographic Code (PSGC) is a systematic classification and coding of geographic areas in the Philippines.

```python
import requests
import csv 

headers = {'Accept': 'application/json', 
           'Content-Type':'application/json'}
```

```python
url = "https://psgc.gitlab.io/api/island-groups/visayas/provinces/"
```

```python
response = requests.request('GET',url,headers= headers , data = {})
visayas = response.json()
```

```python
provinces=[]
for i in range(0,len(visayas)): 
    provinces.append(visayas[i]['name'].upper())

print(provinces)
print(len(provinces))
```

```python
luzon_url = "https://psgc.gitlab.io/api/island-groups/luzon/provinces/"
response = requests.request('GET',luzon_url,headers= headers , data = {})
luzon = response.json()

for i in range(0,len(luzon)): 
    provinces.append(luzon[i]['name'].upper())


print(provinces)
print(len(provinces))
```

# Merging 90_09 and 10_22 area datasets

To merge datasets
1. Delete rows with null locations 
2. Convert location names to uppercase
3. Remove Philippines and Regional values since the study focuses on provincial level 
4. Check if location names are updated 
5. Merge on Location
4. Transform dataset using melt()

```python
def before_merge(dataframe):
    dataframe.drop(dataframe[dataframe['Location'].isnull()==True].index,inplace = True)#delete rows with null locations
    dataframe['Location'] = dataframe['Location'].apply(lambda x: x.upper()) #convert location names to uppercase
    dataframe.drop(dataframe[dataframe['Location'].str.contains('PHILIPPINES|REGION')].index,inplace=True) #remove PH and Regional values
    return dataframe.reset_index(drop=True)
```

```python
ph_area_90_09 = before_merge(ph_area_90_09)
```

```python
ph_area_10_22 = before_merge(ph_area_10_22)
```

```python
#check if location names are updated 

def check_province(list1,list2):
    l2 = set(provinces)
    not_same= [x for x in list1 if x not in l2] #check what provinces are in the dataset that don't have updated names 
    print(not_same)
```

Province names of 1990 - 2009 dataset are up to date based from PSGC 2022 data

```python
check_province(ph_area_90_09['Location'].unique(),provinces)
```

There is still NAN values for '10-'22 area dataset so we will remove those

```python
check_province(ph_area_10_22['Location'].unique(),provinces)
```

```python
ph_area_10_22.drop(ph_area_10_22[ph_area_10_22['Location'] == "NAN"].index,inplace=True)
```

# Compute for yield

```python
merged_ph_data.info()
```

### Convert data type to numeric

```python
def clean_column(col):
    col = col.replace('[^0-9]+', 0, regex=True)
    col = col.astype(float)
    return col
```

```python
merged_ph_data.columns
```

```python
merged_ph_data['Production (mt)'] = clean_column(merged_ph_data['Production (mt)'])
```

```python
merged_ph_data['Area (ha)']  = clean_column(merged_ph_data['Area (ha)'])
```

```python
merged_ph_data.info()
```

```python
merged_ph_data['Yield (mt/ha)'] = merged_ph_data['Production (mt)'] / merged_ph_data['Area (ha)']
```

```python
merged_ph_data = merged_ph_data.reset_index(drop=True)
```
