# Computing for yield 

After merging production and area datasets, check if the data is ready for computation of yield

```python
merged_data.info()
```

The production and area columns are not yet ready for computation since the datatype is object. Hence, it needs to be cleaned then converted into numeric.

### Clean production data 

* remove whitespaces and commas
* convert type to float
* fill null values with 0 (temporary)

```python
merged_data['Production (mt)'] = merged_data['Production (mt)'].str.replace(r"[^0-9.]","", regex=True)
```

```python
merged_data['Production (mt)'] =  merged_data['Production (mt)'].astype(float)
```

```python
merged_data['Production (mt)'].fillna(0,inplace=True)
```

```python
merged_data['Production (mt)'].isnull().sum()
```

### Clean area data

* remove whitespaces and commas
* convert type to float
* fill null values with 0 (temporary)

```python
import numpy as np
```

```python
merged_data['Area (ha)'] = merged_data['Area (ha)'].str.replace(r"[^0-9.]","", regex=True)
merged_data['Area (ha)'] = merged_data['Area (ha)'].replace("", np.NaN)
```

```python
merged_data['Area (ha)'] = merged_data['Area (ha)'].astype(float)
```

```python
merged_data['Area (ha)'].fillna(0,inplace=True)
```

```python
merged_data['Area (ha)'].isnull().sum()
```

```python
merged_data.info()
```

### Compute for yield

```python
merged_data['Yield (mt/ha)'] = merged_data['Production (mt)']/ merged_data['Area (ha)']
```

```python
merged_data.drop(columns=merged_data.columns[0], axis=1, inplace=True)
```

```python
merged_data.head()
```

```python
merged_data.to_csv("Yield Data Final.csv")
```
