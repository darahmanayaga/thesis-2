# Merge area datasets (1990 - 2022)

### Convert location to uppercase for uniformity

```python
area_data_10_21['Location'] = area_data_10_21['Location'].apply(lambda x: x.upper())
```

```python
area_data_90_09['Location'] = area_data_90_09['Location'].apply(lambda x: x.upper())
```

```python
area_data_10_21.head()
```

```python
area_data_90_09.head()
```

```python
list_1 = area_data_10_21['Location'].unique()
list_2 = area_data_90_09['Location'].unique()
list_1.sort()
list_2.sort()

common = list(set(list_1) & set(list_2))


for x in list_2:
    if x not in common: 
        print(x)

```

### Province names of ARMM and Compostela Valley must be changed to the updated names

```python
area_data_90_09['Location'] = area_data_90_09['Location'].replace({'AUTONOMOUS REGION IN MUSLIM MINDANAO (ARMM)':'BANGSAMORO AUTONOMOUS REGION IN MUSLIM MINDANAO (BARMM)', 
'COMPOSTELA VALLEY': 'DAVAO DE ORO'})
```

Chech if changes were applied to the dataset

```python
list_1 = area_data_10_21['Location'].unique()
list_2 = area_data_90_09['Location'].unique()
list_1.sort()
list_2.sort()

common = list(set(list_1) & set(list_2))


for x in list_2:
    if x not in common: 
        print(x)

```

```python
area_data = pd.concat([area_data_90_09, area_data_10_21])
```

```python
area_data
```

```python
area_data.to_csv("Area Data Final.csv")
```
