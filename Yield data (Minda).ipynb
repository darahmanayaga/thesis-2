### Convert data type of production and area to numeric

```python
import numpy as np
def clean_column(col):
    col = col.replace('[^0-9.]','', regex=True)
    col = col.astype(float)
    return col
```

```python
merged_min_data['Production (mt)'] = clean_column(merged_min_data['Production (mt)'])
```

```python
merged_min_data['Area (ha)'] = clean_column(merged_min_data['Area (ha)'])
```

```python
merged_min_data
```

There are still null values for area after cleaning the columns so we will remove those

```python
merged_min_data.info()
```

```python
# remove NAN
merged_min_data.drop(merged_min_data[merged_min_data['Area (ha)'].isnull()==True].index,inplace=True)
```

```python
merged_min_data.info()
```

### Compute for yield

```python
merged_min_data['Yield (mt/ha)'] = merged_min_data['Production (mt)'] / merged_min_data['Area (ha)']
```
