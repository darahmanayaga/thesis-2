```python
yield_data=pd.concat([ZAMBOANGA_DEL_NORTE,ZAMBOANGA_DEL_SUR, ZAMBOANGA_SIBUGAY, CITY_OF_ZAMBOANGA, BUKIDNON, CAMIGUIN, LANAO_DEL_NORTE, MISAMIS_OCCIDENTAL, MISAMIS_ORIENTAL, DAVAO_DE_ORO, DAVAO_DEL_NORTE, DAVAO_DEL_SUR, DAVAO_OCCIDENTAL, DAVAO_ORIENTAL, CITY_OF_DAVAO, COTABATO, SARANGANI, SULTAN_KUDARAT, SOUTH_COTABATO, AGUSAN_DEL_NORTE, AGUSAN_DEL_SUR, DINAGAT_ISLANDS, SURIGAO_DEL_NORTE, SURIGAO_DEL_SUR, BASILAN, LANAO_DEL_SUR, MAGUINDANAO, SULU, TAWI_TAWI],ignore_index = True)
```

```python
yield_data = yield_data.drop(['Unnamed: 0'],axis = 1)
```

```python
weather_data.rename(columns = {'year':"Year"},inplace=True)
```

Check if the names of provinces are the same for both datasets 

```python
list_1 = weather_data['Location'].unique()
list_2 = yield_data['Location'].unique()
list_1.sort()
list_2.sort()

common = list(set(list_1) & set(list_2))


for x in list_2:
    if x not in common: 
        print(x)
```

Rename zamboanga city to city of zamboanga

```python
weather_data['Location'] = weather_data['Location'].replace({'ZAMBOANGA CITY':'CITY OF ZAMBOANGA'})
```

Merge datasets on year and location columns

```python
yield_weather = pd.merge(yield_data,weather_data, how = 'inner', on =['Year','Location'])
```
